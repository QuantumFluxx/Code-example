## 4. Напишите 3 JTBD для сервиса по доставке продуктов «Самокат»

1. Когда у меня нет возможности пойти в магазин, я хочу заказать доставку еды, чтобы это было быстро и до двери, так я могу потратить свое время на личные дела вместо похода в магазин.
2. Когда я захожу в приложение по доставке еды, я хочу, чтобы интерфейс был интуитивно понятен и я быстро смог найти и заказать необходимые продукты, так я останусь доволен и не захочу перейти к другому сервису, в котором все удобно и понятно.
3. Когда я делаю заказ продуктов, я хочу, чтобы все необходимое было доставлено без ошибок и вовремя, так я не буду тратить свое время на разрешение неприятной ситуации.
